```julia
meanstdm = DataFrame()
meanstdm[:mean] = describe(results)[:mean]
meanstdm[:stdm] = [stdm(results[:vxv100k], mean(results[:vxv100k])),
    stdm(results[:vxv200k], mean(results[:vxv200k])),
    stdm(results[:vxv300k], mean(results[:vxv300k])),
    stdm(results[:vxv400k], mean(results[:vxv400k])),
    stdm(results[:vxv500k], mean(results[:vxv500k])),
    stdm(results[:mxv1k], mean(results[:mxv1k])),
    stdm(results[:mxv2k], mean(results[:mxv2k])),
    stdm(results[:mxv3k], mean(results[:mxv3k])),
    stdm(results[:mxv4k], mean(results[:mxv4k])),
    stdm(results[:mxv5k], mean(results[:mxv5k]))
    ]
meanstdm[:sizes] = [100000, 200000, 300000, 400000, 500000, 1000, 2000, 3000, 4000, 5000]
meanstdm[:tryby] = ["vxv","vxv","vxv","vxv","vxv","mxv","mxv","mxv","mxv","mxv"]
show(meanstdm)
```

```julia
plot(meanstdm[:sizes], meanstdm[:mean], group=meanstdm[:tryby], colour = [:red :blue], layout=2, yerr=meanstdm[:stdm])
```
