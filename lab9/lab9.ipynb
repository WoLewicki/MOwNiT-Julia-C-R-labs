# Zadanie symulacji uk≈Çadu grawitacyjnego

```julia
function my_gravity(dt, r, v, t, results)
    a = [0.0,0.0,0.0]
    for x=0:dt:t
        scalR = r[1]^2 + r[2]^2 + r[3]^2
        for i=1:3
            a[i] = -r[i] / (scalR *sqrt(scalR))
            r[i] += v[i]*dt
            v[i] += a[i]*dt
            push!(results, r[i])
        end
    end
end
```

```julia
@manipulate for rx = 0.0:0.01:5.0, ry = 0.0:0.01:5.0, rz = 0.0:0.01:5.0, vx = 0.0:0.01:5.0,
vy = 0.0:0.01:5.0, vz = 0.0:0.01:5.0, timer = 0.0:0.01:30.0, dtime = 0.0001:0.0001:1.0
dt = dtime
r = [rx,ry,rz]
v = [vx, vy, vz]
t = timer
results = []
my_gravity(dt,r,v,t, results)
resx = []
resy = []
resz = []
for i=1:3:(Int(floor(t/dt)))*3
    push!(resx, results[i])
    push!(resy, results[i+1])
    push!(resz, results[i+2])
end
plot(resx, resy, resz)
scatter!([0],[0],[0])
end
```

```julia
function ode_gravity1(dv, v, r, p, t)
    #scalar = (sqrt(r[1]^2+r[2]^2+r[3]^2))
    dv[1] -= r[1]/((sqrt(r[1]^2+r[2]^2+r[3]^2))^3)
    dv[2] -= r[2]/((sqrt(r[1]^2+r[2]^2+r[3]^2))^3)
    dv[3] -= r[3]/((sqrt(r[1]^2+r[2]^2+r[3]^2))^3)
end
```

```julia
function ode_gravity2(dr, v, r, p, t)
    dr[1] += v[1]
    dr[2] += v[2]
    dr[3] += v[3]
end
```

```julia
@manipulate for rx = 0.0:0.01:5.0, ry = 0.0:0.01:5.0, rz = 0.0:0.01:5.0, vx = 0.0:0.01:5.0,
vy = 0.0:0.01:5.0, vz = 0.0:0.01:5.0, timer = 0.0:0.01:5.0, dtime = 0.0001:0.0001:1.0
dt = dtime
r0 = [rx,ry,rz]
v0 = [vx, vy, vz]
tspan = timer
prob = DynamicalODEProblem(ode_gravity1, ode_gravity2, v0, r0, tspan)
sol = solve(prob, IRKN4(), dt = dtime)
u1 = []
u2 = []
u3 = []
for i=1:size(sol.u)[1]
    push!(u1, sol.u[i][4])
    push!(u2, sol.u[i][5])
    push!(u3, sol.u[i][6])
end
scatter(u1, u2, u3)
scatter!([0],[0],[0])
end
```
